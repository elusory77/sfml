```python
# Попробовать следующие "черные ящики": интерфейс одинаковый 
# Постепенно мы узнаем, что они делают а сейчас учимся понимать какой работает качественнее
#     fit, 
#     predict, 
#     predict_proba
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

model=DecisionTreeClassifier()
model.fit(x_train,y_train)
predict=model.predict(x_test)
prediction_probs = model.predict_proba(x_test)

model2=RandomForestClassifier()
model2.fit(x_train,y_train)
predict2=model2.predict(x_test)
prediction_probs2 = model2.predict_proba(x_test)

model3=LogisticRegression()
model3.fit(x_train,y_train)
predict3=model3.predict(x_test)
prediction_probs3 = model3.predict_proba(x_test)
```

```python
# Посчитать метрики стандартные
# accuracy, precision, recall
from sklearn.metrics import accuracy_score, precision_score, recall_score

print('DecisionTreeClassifier')
p1=precision_score(y_test,predict)
print('precision',precision_score(y_test,predict))
r1=recall_score(y_test,predict)
print('recall',recall_score(y_test,predict))
a1=accuracy_score(y_test,predict)
print('accuracy_score',accuracy_score(y_test,predict))

print('\nRandomForestClassifier')
p2=precision_score(y_test,predict2)
print('precision',p2)
r2=recall_score(y_test,predict2)
print('recall',r2)
a2=accuracy_score(y_test,predict2)
print('accuracy_score',a2)

print('\nLogisticRegression')
p3=precision_score(y_test,predict3)
print('precision',p3)
r3=recall_score(y_test,predict3)
print('recall',r3)
a3=accuracy_score(y_test,predict3)
print('accuracy_score',a3)
```

```python
# Потроить roc-кривые всех можелей на одном графике
# Вывести roc_auc каждой моделе
# Подпишите оси и линии
from sklearn.metrics import roc_auc_score, roc_curve

fpr,tpr,tresholds = roc_curve(y_test,prediction_probs[:,1])
plt.figure(figsize=(10,10))
plt.plot(fpr,tpr)

plt.plot([0,1],[0,1])
plt.ylabel('tpr')
plt.xlabel('fpr')
plt.grid('True')
plt.title('ROC-curve')
plt.legend('DecisionTreeClassifier')

print('ROC-auc для DecisionTreeClassifier',roc_auc_score(y_test,prediction_probs[:,1]))

fpr2,tpr2,tresholds2 = roc_curve(y_test,prediction_probs2[:,1])
plt.figure(figsize=(10,10))
plt.plot(fpr2,tpr2)

plt.plot([0,1],[0,1])
plt.ylabel('tpr')
plt.xlabel('fpr')
plt.grid('True')
plt.title('ROC-curve')
plt.legend('RandomForestClassifier')

print('ROC-auc для RandomForestClassifier',roc_auc_score(y_test,prediction_probs2[:,1]))

fpr3,tpr3,tresholds3 = roc_curve(y_test,prediction_probs3[:,1])
plt.figure(figsize=(10,10))
plt.plot(fpr3,tpr3)

plt.plot([0,1],[0,1])
plt.ylabel('tpr')
plt.xlabel('fpr')
plt.grid('True')
plt.title('ROC-curve')
plt.legend('LogisticRegression')

print('ROC-auc для LogisticRegression',roc_auc_score(y_test,prediction_probs3[:,1]))
```
