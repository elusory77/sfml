```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import roc_auc_score
```

```python
df_train = pd.read_csv('C://PML//Lesson3//all//train.csv')
```

```python
df_test = pd.read_csv('C://PML//Lesson3//all//test.csv')
```

```python
df_train.loc[:, 'sample'] = 'train'
df_test.loc[:, 'sample'] = 'test'
```

```python
#Обьединяем
df = df_train.append(df_test).reset_index(drop=True)
```

```python
def preproc(df_init): 
    df_preproc = df_init.copy()

  # Удалили признаки
    df_preproc = df_preproc.drop(['contact'], axis=1)

    from sklearn.preprocessing import LabelEncoder
   #Просто перебрал по столбцам 
    LE = LabelEncoder()
    LE.fit(df_preproc.job)
    df_preproc['job']=LE.transform(df_preproc.job)
    
    LE2 = LabelEncoder()
    LE2.fit(df_preproc.marital)
    df_preproc['marital']=LE2.transform(df_preproc.marital)
    
    LE3 = LabelEncoder()
    LE3.fit(df_preproc.education)
    df_preproc['education']=LE3.transform(df_preproc.education)    

    LE4 = LabelEncoder()
    LE4.fit(df_preproc.default)
    df_preproc['default']=LE4.transform(df_preproc.default) 
    
    LE6 = LabelEncoder()
    LE6.fit(df_preproc.housing)
    df_preproc['housing']=LE6.transform(df_preproc.housing)     
    
    LE7 = LabelEncoder()
    LE7.fit(df_preproc.loan)
    df_preproc['loan']=LE7.transform(df_preproc.loan) 
    
    LE8 = LabelEncoder()
    LE8.fit(df_preproc.month)
    df_preproc['month']=LE8.transform(df_preproc.month) 
    
    LE9 = LabelEncoder()
    LE9.fit(df_preproc.day_of_week)
    df_preproc['day_of_week']=LE9.transform(df_preproc.day_of_week)
    
    LE10 = LabelEncoder()
    LE10.fit(df_preproc.poutcome)
    df_preproc['poutcome']=LE10.transform(df_preproc.poutcome)
    
    return df_preproc
```

```python
df_preproc = df.pipe(preproc)
```

```python
df_trn_prep = df_preproc.query('sample == "train"').drop(['sample'], axis=1)
df_test_prep = df_preproc.query('sample == "test"').drop(['sample'], axis=1)

#X, y = df_trn_prep.iloc[:, :-1].values, df_train.iloc[:, -1].values
```

```python
X_train, X_test = df_trn_prep.drop(['target','_id'], axis=1), df_test_prep.drop(['target','_id'], axis=1)
y_train = df_trn_prep.iloc[:, -1].values
y_test = df_test_prep.iloc[:, -1].values
```

```python
for k in range(55,60):
    model = Pipeline([
        ('scaler', StandardScaler()),
        ('knn', KNeighborsClassifier(n_neighbors=k))
    ])
    
    model.fit(X_train, y_train)
    y_hat = model.predict(X_test)
    y_hat_proba = model.predict_proba(X_test, )
    print('k=',k,'roc_auc=',roc_auc_score(y_test, y_hat_proba[:, 1]))
  #Лучшая метрика - k= 55 roc_auc= 0.9173900814770245
```
